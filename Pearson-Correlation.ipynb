# <span style="color:lightgreen; font-size:18px;">Pearson Correlation</span>

Concept

Pearson Correlation Coefficient (r) → mengukur kekuatan & arah hubungan linear antara dua variabel numerik.

Nilai r:

+1 → strong positive correlation

-1 → strong negative correlation

0 → no correlation

p-value → seberapa yakin kita terhadap hasil korelasi (statistical significance):

p < 0.001 → very strong certainty

0.001 ≤ p < 0.05 → moderate certainty

0.05 ≤ p < 0.1 → weak certainty

p ≥ 0.1 → no certainty

⚠️ Catatan: korelasi ≠ kausalitas (correlation does not imply causation).

```python
import pandas as pd
from scipy import stats

# Sample dataset
data = {
    "horsepower": [100, 200, 150, 300, 250, 120, 180, 220],
    "price": [15000, 28000, 20000, 40000, 35000, 16000, 22000, 30000]
}

df = pd.DataFrame(data)

# Pearson correlation
r, p = stats.pearsonr(df["horsepower"], df["price"])

print("Correlation coefficient (r):", r)
print("p-value:", p)

# Interpretation
if p < 0.001:
    certainty = "strong certainty"
elif p < 0.05:
    certainty = "moderate certainty"
elif p < 0.1:
    certainty = "weak certainty"
else:
    certainty = "no certainty"

print(f"Interpretation: r={r:.2f}, p={p:.5f} → {certainty}")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix with Pearson
corr_matrix = df.corr(method="pearson")

plt.figure(figsize=(6,4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", center=0)
plt.title("Pearson Correlation Heatmap")
plt.show()
```
